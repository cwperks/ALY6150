```python
#3 Controlling for confounding variables


df_filtered = df[(df['Type of Admission'] == 'Trauma') & (df['Department'] == 'gynecology') & (df['Hospital_region_code'] == 'X') & (df['Ward_Type'] == 'R') & (df['Hospital_type_code'] == 'a')]




```

```python
df["Severe"] = df['Severity of Illness'] == 'Extreme'
```

```python
df.replace({'Severe': {True: 1, False: 0}}, inplace=True)
```

```python
logit_df = df.drop(columns=['Severity of Illness'], axis=1)
```

```python
logit_df = pd.get_dummies(logit_df, columns=['Department', 'Type of Admission', 'Hospital_type_code', 'Hospital_region_code', 'Ward_Type', 'Ward_Facility_Code'], drop_first=True)
```

```python
df["Age"].str.split("-", 1)
logit_df["Length_of_stay"].value_counts()
```

```python
logit_df.replace({'Length_of_stay': {"More than 100 Days": "100-999"}}, inplace=True)
```

```python
logit_df[['Low Age', 'High Age']] = logit_df["Age"].str.split("-", 1, expand=True)
logit_df['Low Age'] = pd.to_numeric(logit_df['Low Age'])
logit_df['High Age'] = pd.to_numeric(logit_df['High Age'])

logit_df[['Low LOS', 'High LOS']] = logit_df["Length_of_stay"].str.split("-", 1, expand=True)
logit_df['Low LOS'] = pd.to_numeric(logit_df['Low LOS'])
logit_df['High LOS'] = pd.to_numeric(logit_df['High LOS'])
```

```python
logit_df['Avg Age'] = (logit_df['High Age'] + logit_df['Low Age']) / 2
logit_df['Avg LOS'] = (logit_df['High LOS'] + logit_df['Low LOS']) / 2
```

```python
logit_df.drop(columns=['Low Age', 'High Age', 'Age', 'Low LOS', 'High LOS', 'Length_of_stay'], inplace=True, errors='ignore')
```

```python
from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler()

logit_df[['patientid', 'Admission_Deposit', 'Hospital_code', 'Available Extra Rooms in Hospital', 'Bed Grade', 'Visitors with Patient', 'City_Code_Hospital', 'City_Code_Patient', 'Avg Age', 'Avg LOS']] = scaler.fit_transform(logit_df[['patientid', 'Admission_Deposit', 'Hospital_code', 'Available Extra Rooms in Hospital', 'Bed Grade', 'Visitors with Patient', 'City_Code_Hospital', 'City_Code_Patient', 'Avg Age', 'Avg LOS']])


```

```python
logit_df.head().T
```

```python
logit_df.fillna(logit_df.median())
```

```python
logit_df.isna().sum()
```

```python
logit_df.drop(columns=['Bed Grade', 'City_Code_Patient'], inplace=True, errors='ignore')
```

```python
logit_df.describe(include='all').T
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(logit_df[logit_df.columns[~logit_df.columns.isin(['Severe'])]], logit_df['Severe'], test_size=0.25, random_state=0)

```

```python
x_train.head()
```

```python
from sklearn import datasets, linear_model

model1 = linear_model.LogisticRegression()

model1.fit(x_train, y_train)
```

```python
model1.predict(x_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix, accuracy_score

print(f"Accuracy: {accuracy_score(y_test, model1.predict(x_test))}")

plot_confusion_matrix(model1, x_test, y_test)
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

y_score = model1.predict_proba(x_test)

fpr = []
tpr = []
roc_auc = []
fpr, tpr, _ = roc_curve(y_test, y_score[:,1])
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f"Model1 ROC curve of class (area = {roc_auc:.2f})")
    
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([-0.05, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic for Logistic Regression')
plt.legend(loc="lower right")
plt.show()
```
