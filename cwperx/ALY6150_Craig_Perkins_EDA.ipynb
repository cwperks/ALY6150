```python
import pandas as pd
import seaborn
import matplotlib
```

```python
df = pd.read_csv('./BPHC_COVID_download/Boston_COVID_database_download_2022_11_04.csv')
```

```python
df.head(10)
```

```python
df.sample(5).T
```

```python
df.drop(columns='Chart_info', inplace=True, errors='ignore')
```

```python
df.describe(include='all')
```

```python
positive_tests = df[df['Title'] == 'COVID-19 Positive Tests']
```

```python
lines = positive_tests.plot.line(x='Category1', y='Value', title='COVID-19 Positive Tests for Boston Area (BPHC)')
lines.set_xlabel("Number of Positive Tests")
lines.set_ylabel("Date")
```

```python
race = df[df['Title'] == 'COVID-19 Cases by Race/Ethnicity (Based on 88.5 Percent of \nTotal Cases in Boston with Known race/Ethnicity)']
```

```python
race = race.set_index('Category1')
```

```python
race
```

```python
plot = race.plot.pie(y='Value', title='COVID-19 Percentage by Race')
```

```python
gender = df[df['Title'] == 'COVID-19 Rates by City and Gender']
```

```python
gender
```

```python
ax = gender.plot.bar(x='Category1', y='Value', rot=0, title='COVID-19 rate per 100k by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('COVID-19 rate per 100k')
```
